```python
import pandas as pd
marks = [50000, 55000, 60000, 65000, 70000, 75000, 80000, 85000, 90000, 95000, 100000, 105000, 110000, 115000, 120000]
df = pd.DataFrame(marks, columns=['Salary'])
df['Employee_ID'] = range(1, len(df) + 1)
print(df)
```

```python
print("First 5 rows:\n", df.head())
print("Last 5 rows:\n", df.tail())

print("Mean:", df['Salary'].mean())
print("Median:", df['Salary'].median())
print("Mode:", df['Salary'].mode()[0])
print("Standard Deviation:", df['Salary'].std())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
hist_plot = sns.histplot(df['Salary'], kde=True, color='skyblue')

for p in hist_plot.patches:
    if p.get_height() > 0:
        hist_plot.annotate(f'{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                           ha='center', va='center', xytext=(0, 5), textcoords='offset points')

plt.title('Distribution of Salary', fontsize=16)
plt.xlabel('Salary', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.xticks(np.arange(0, 100000, 20000), fontsize=12)
plt.grid(True, linestyle='--', linewidth=0.7)
plt.show()
```

```python
sns.boxplot(x=df['Salary'], color='Yellow')
plt.title('Box Plot of Salary')
plt.xticks(np.arange(0, 100000, 20000))
plt.show()
```

```python
def categorize_salary(Salary):
    if Salary >= 120000:
       return 'A'
    elif Salary >= 100000:
       return 'B'
    elif Salary >= 80000:
       return 'C'
    elif Salary >= 60000:
       return 'D'
    else:
       return 'E'

df['Persantage'] = df['Salary'].apply(categorize_salary)
persantage_counts = df['Persantage'].value_counts()
plt.pie(persantage_counts, labels=persantage_counts.index, autopct='%1.1f%%', colors=['gold', 'lightcoral',
'lightskyblue', 'lightgreen', 'red'])
plt.title('persantage Distribution')
plt.show()
```

```python
sns.kdeplot(df['Salary'], fill=True, color='blue')
plt.title('KDE Plot of Salary')

plt.xlabel('Salary')
plt.ylabel('Density')
plt.show()
```

```python
from scipy import stats
import statsmodels.api as sm
sm.qqplot(df['Salary'], line='s')
plt.title('Q-Q Plot of Salary')

plt.show()
```
